```python
import torch
from utils.model_utils import load_generator, generate
from utils.visualize import *
from utils.FragmentDataset import FragmentDataset
```

```python
RESOLUTION = 32
path_checkpoint = "./checkpoints/runs_cos_stable/model_epoch_100.pth"
G = load_generator(path_checkpoint, RESOLUTION)
```

```python
dataset = FragmentDataset('./data', 'test', dim_size=RESOLUTION, use_cache=False)
N = dataset.__len__()
ID = 655
assert ID < N, f"ID {ID} exceeds dataset length {N}"

fractures = dataset.__getfractures__(ID)
print(fractures)
frag_id = fractures[2]
print(ID, frag_id)

gt, _, _, _ = dataset.__read_vox__(dataset.vox_files[ID])
frag, _ = dataset.__select_fragment_specific__(gt, [frag_id])

plot_frag(gt, './test_visualize_data/output/gt.png')
plot_frag(frag, './test_visualize_data/output/frag.png')
```

```python
result, _ = generate(G, frag)
result = result.squeeze()
result = result > 0.5
plot(result, './test_visualize_data/output/result.png')
```
